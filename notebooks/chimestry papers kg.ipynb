```python
import json

with open('chimestry_papers_processed.json') as json_file:
    chimestry_papers = json.load(json_file)
import pandas as pd 
data = pd.read_excel('./data/wos_um6p.xls') 
import pandas as pd 
umls_tui = pd.read_csv('umls_types.csv')
def get_umls_label(code):
    return umls_tui[umls_tui['tui']==code]['label'].values[0]
```

```python
# only Laboratory Procedure and chemical substances types
graph_node_types = ['T109','T059','T197','T196','T167','T116']  
```

```python

## get unique keywords 
keywords_unique = []
for key in chimestry_papers :
    keywords = chimestry_papers[key]['keywords']
    for keyword in keywords :
        if keyword['label'] not in keywords_unique:
            keywords_unique.append(keyword['label'])
```

```python
keyword_nodes = []

for keyword_label in keywords_unique :
    d=[]
    papers_count=[]
    for key in chimestry_papers :
        keywords = chimestry_papers[key]['keywords']
        for keyword in keywords :
            if keyword['label']==keyword_label:
                papers_count.append(key) 
                type_= keyword['types']
    if type_ in  graph_node_types :           
        keyword_nodes.append([keyword_label,type_,len(papers_count),papers_count])


              
    
```

```python
keyword_nodes
```

```python
## edges 
keyword_edges = []
for node_i in keyword_nodes:
    for node_j in keyword_nodes :
        if node_i==node_j:
            pass
        else:
            node_i_papers = node_i[3]
            node_j_papers = node_j[3]
            inter = list(set(node_i_papers) & set(node_j_papers))
            if len(inter)!=0:
                keyword_edges.append([node_i[0],node_j[0],len(inter)])
            
```

```python
keyword_edges
```

```python
edges_pd = pd.DataFrame(keyword_edges,columns=['source','target','weight'])
nodes_pd = pd.DataFrame([[node[0],get_umls_label(node[1]),node[2]]for node in keyword_nodes],columns=['id','type','paper_count'])
```

```python
edges_pd
```

```python
## save data for Gephy visualization
edges_pd.to_csv('edges_chim.csv',index=False)
nodes_pd.to_csv('nodes_chim.csv',index=False)
```

```python

## filtred view of graph only take one edge with max co occurence !
def get_max_edge(node):
    new_d = edges_pd[(edges_pd['source']==node) |(edges_pd['source']==node)]
    try :
        return list(new_d.sort_values(by=['weight'],ascending=False).values[0])
    except : 
        return False
get_max_edge('liquids')   
```

```python
edges_max = []
for node in nodes_pd['id'].values:
    #print(node)
    edge =  get_max_edge(node)
    if edge:
        edges_max.append(edge)
```

```python
# df for max edge graph view
edges_pd_max = pd.DataFrame(edges_max,columns=['source','target','weight'])
```
