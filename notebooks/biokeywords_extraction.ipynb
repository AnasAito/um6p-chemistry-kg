```python
import scispacy
import scispacy
import spacy
from spacy import displacy  
## nlp pipeline for NER task
nlp = spacy.load("en_core_sci_lg")
## linker to umls database for keywords labeling
nlp.add_pipe("scispacy_linker", config={ "linker_name": "umls"})
linker = nlp.get_pipe("scispacy_linker")
```

```python
# test
text = data[data['Research Areas']==area][['Abstract']].values[0][0]
doc = nlp(text)

displacy_image = displacy.render(doc, jupyter = True, style = 'ent')
```

```python
def get_umls_meta(entity):
    #print(entity._.kb_ents)
    if entity._.kb_ents!=[]:
            umls_ent = entity._.kb_ents[0]
            types = linker.kb.cui_to_entity[umls_ent[0]].types
            canonical = linker.kb.cui_to_entity[umls_ent[0]].canonical_name
            return {'label':str(entity),'canonical':canonical,'types':types}
    
    
```

```python
def extract_entities(text):
    doc = nlp(text)
    L = [get_umls_meta(ent) for ent in doc.ents] 
    return [x for x in L if x is not None]
```

```python
chimestry_papers = {}
for paper in data[data['Research Areas']==area][['UT (Unique WOS ID)','Article Title','Abstract','Publication Year','Document Type']].values :
    
    keywords = extract_entities(paper[2])
    #print(keywords)
    #print('---------------')
    chimestry_papers[paper[0]] = {'title':paper[1],'year':paper[3],'type':paper[4],'keywords' :keywords}
```

```python
# save raw data 
import json
with open('chimestry_papers.json','w') as f:
    json.dump(chimestry_papers, f, ensure_ascii=False, indent=4)
```
