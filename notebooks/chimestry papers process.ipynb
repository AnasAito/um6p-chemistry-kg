```python
def get_keywords_by_code(data_dict,code_list):
    k=[]
    paper_keys = []
    for key in data_dict.keys():
        ## get keywords 
        keywords = data_dict[key]['keywords']
        for keyword in keywords :
            test = list(set(code_list) & set(keyword['types']))!=[]
            if test :
                keyword_ = keyword
                keyword_['paper_id'] = key
                paper_keys.append(key)
                k.append(keyword_)
    keywords_code = list({v['label']:v for v in k}.values())
    return keywords_code,{'unique_papers_ratio':round(100*(len(set(paper_keys))/len(data_dict)),1),'keywords_count':len(keywords_code)}
# test on labo procedure code
labo_proce,logs = get_keywords_by_code(chimestry_papers,['T059'])
logs
```

```python
# util to get keywords with same canonical form 
def get_labels(data_dict,cano):
    k=[]
    paper_keys = []
    for key in data_dict.keys():
        ## get keywords 
        keywords = data_dict[key]['keywords']
        for keyword in keywords :
      
            if keyword['canonical']==cano :


                k.append(keyword['label'].lower())

    return list(set(k))
get_labels(chimestry_papers, 'Reduced')    
```

```python
cano = []
for key in chimestry_papers.keys():
        keywords = chimestry_papers[key]['keywords']
        for keyword in keywords :
            cano.append(keyword['canonical'])
cano_list = list(set(cano))
can_multiple = []
for can in cano_list:
    can_labels = get_labels(chimestry_papers, can)   
    if len(can_labels)>1:
        can_multiple.append((can,can_labels))
        
```

```python
label_to_cano = [can[0] for can in can_multiple if can[0] not in no_relabel]
```

```python
# relabel 
label_to_cano
# code delete  :delete codes related to medical usecases 
code_delete = ['T121','T122','T126','T121','T019','T123','T125','T131','T130','T195','T129','T127']
```

```python
# update dict 
def update(paper):
    ## relabel 
    keywords = paper['keywords']
    keywords = list({v['label']:v for v in keywords}.values())
    keywords_new = []
    for keyword in keywords : 
        # label task
        is_relabel = (keyword['canonical'] in label_to_cano) 
        update_keyword = keyword
        if is_relabel :
            
            update_keyword['label'] = keyword['canonical']
            #keywords_new.append(update_keyword)
        else : 
            pass
        
        ## code task 
        codes = keyword['types']
        
        if len(codes)>1:
            new_codes = [code for code in codes if ( code not in code_delete) ]
        else :
            new_codes= codes
        if   new_codes==[]:
            update_keyword['types']=''
        else :
            update_keyword['types']=new_codes[0]
        ## new keyword list  
        keywords_new.append(update_keyword) 
        #print(len(keywords_new))
        
    return {'title':paper['title'],'year':paper['year'],'type':paper['type'],'keywords':keywords_new}
        
new_chimestry_papers = {k: update(v) for k, v in chimestry_papers.items()}
```
